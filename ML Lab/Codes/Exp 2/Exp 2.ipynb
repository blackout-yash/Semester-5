```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
data = pd.read_csv("tips.csv")
data.head()
```

```python
plt.scatter(data['day'], data['tip'])
plt.title('scatterplot')
plt.xlabel('day')
plt.ylabel('tip')
```

```python
plt.scatter(data['day'], data['tip'], c = data['total_bill'], s = data['size'])
plt.xlabel('day')
plt.ylabel('tip')

plt.colorbar()
plt.show()
```

```python
plt.bar(data['day'], data['tip'])
plt.title('scatterplot')
plt.xlabel('day')
plt.ylabel('tip')

plt.show()
```

```python
plt.hist(data['day'])
plt.title('histogram')
plt.show()
```

```python
sn.scatterplot(x = 'day', y = 'tip', data = data)
plt.show()
```

```python
sn.scatterplot(x = 'day', y = 'tip', data = data, hue = 'sex')
plt.show()
```

```python
sn.lineplot(x = 'day', y = 'tip', data = data)
plt.show()
```

```python
sn.barplot(x = 'day', y = 'tip', data = data, hue = 'sex')
plt.show()
```

```python
sn.histplot(x = 'day', data = data, hue = 'size', kde = True)
plt.show()
```
